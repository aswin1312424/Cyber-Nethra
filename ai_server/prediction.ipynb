```python
import pickle
```

```python
with open("url_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("vectorizer.pkl", "rb") as f:
    vectorizer = pickle.load(f)
```

```python
def predict_url(url):
    x = vectorizer.transform([url])
    pred = model.predict(x)[0]

    return "Safe" if pred == 1 else "Malicious"
```

```python
print(predict_url("yourbittorrent.com/?q=anthony-hamilton-soulife"))
```

```python
proba = model.predict_proba(vectorizer.transform(["https://google.com"]))
print(proba)
```

```python
proba = model.predict_proba(vectorizer.transform(["yourbittorrent.com/?q=anthony-hamilton-soulife"]))
print(proba)
```
